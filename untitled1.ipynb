```python
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
# Membuat Database Mentor Dummy
mentors_df = pd.read_csv("Large_Dummy_Mentor_Database.csv")
mentors_df.tail()
```

```python
plt.figure(figsize=(12, 6))
top_10_mentors = mentors_df.head(10)

# Membuat visualisasi
plt.figure(figsize=(12, 6))
plt.bar(top_10_mentors["Mentor Name"], top_10_mentors["Expertise"], color="skyblue")
plt.xlabel("Mentor Name", fontsize=12)
plt.ylabel("Match Score", fontsize=12)
plt.title("Top 10 Mentor Match Scores", fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
